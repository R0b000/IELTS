```python
import nest_asyncio
from pyngrok import ngrok
import uvicorn
from fastapi import FastAPI, HTTPException, WebSocket
from pydantic import BaseModel
from transformers import AutoTokenizer, AutoModelForCausalLM, AsyncTextIteratorStreamer
from threading import Thread
import torch

# Apply nested asyncio for ngrok compatibility
nest_asyncio.apply()

# Initialize FastAPI app
app = FastAPI()

# Load the tokenizer and model
MODEL_NAME = "R0b000/Fine_tuned_Mistral_IELTS"  # Replace with your model's path
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME).to("cuda")  # Ensure GPU support

# Define the request body structure
class StreamRequest(BaseModel):
    prompt: str

# WebSocket endpoint for real-time streaming
@app.websocket("/stream/")
async def stream(websocket: WebSocket):
    await websocket.accept()
    try:
        # Receive the prompt from the client
        data = await websocket.receive_json()
        prompt = data.get("prompt")
        if not prompt:
            await websocket.close(code=1003, reason="Invalid input: 'prompt' is required.")
            return

        print(f"Received WebSocket request with prompt: {prompt}")

        # Define the formatted PROMPT
        PROMPT = f"""
        In this task, you are required to evaluate an IELTS Writing Task 2 essay. Consider the following four criteria and provide a detailed assessment for each, along with a suggested band score:

        ## Task Achievement:
        - Evaluate how well the candidate has addressed the given task.
        - Assess the clarity and coherence of the response in presenting ideas.
        - Identify if the candidate has fully covered all parts of the task and supported arguments appropriately.
        - Suggested Band Score (Task Achievement): [Insert Score]

        ## Coherence and Cohesion:
        - Assess the overall organization and structure of the essay.
        - Evaluate the use of linking devices to connect ideas and paragraphs.
        - Identify if there is a logical flow of information.
        - Suggested Band Score (Coherence and Cohesion): [Insert Score]

        ## Lexical Resource (Vocabulary):
        - Examine the range and accuracy of vocabulary used in the essay.
        - Point out specific mistakes in vocabulary, such as inaccuracies or overuse of certain words and suggest modified versions or alternatives for the identified mistakes.
        - Assess the appropriateness of vocabulary for the given context.
        - Suggested Band Score (Lexical Resource): [Insert Score]

        ## Grammatical Range and Accuracy:
        - Evaluate the variety and complexity of sentence structures.
        - Point out specific grammatical errors, such as incorrect verb forms or sentence construction and suggest modified versions or corrections for the identified mistakes.
        - Examine the use of punctuation and sentence formation.
        - Suggested Band Score (Grammatical Range and Accuracy): [Insert Score]

        ## Overall Band Score:
        - Provide an overall band score for the essay, considering the holistic performance across all criteria.
        - Consider the synergy of the essay in meeting the task requirements cohesively.
        - Suggested Overall Band Score: [Insert Score]

        ## Feedback and Additional Comments:
        - Provide constructive feedback highlighting specific strengths and areas for improvement.
        - Suggest strategies for enhancement in weaker areas.

        ## Input:
        {prompt}

        ## Evaluation:
        """

        # Prepare input for the model
        inputs = tokenizer([PROMPT], return_tensors="pt").to("cuda")
        input_ids = inputs['input_ids']  # Get the tensor from the tokenizer output

        # Initialize the async streamer
        streamer = AsyncTextIteratorStreamer(tokenizer, skip_prompt=True, skip_special_tokens=True)

        # Start the generation in a separate thread
        generation_kwargs = dict(
            input_ids=input_ids,  # Use the input_ids tensor
            max_new_tokens=1024,  # Adjust token limit as needed
            temperature=0.8,     # Adjust sampling temperature
            do_sample=True,
            streamer=streamer,
        )
        thread = Thread(
            target=model.generate,
            kwargs=generation_kwargs,
        )
        thread.start()

        # Stream the generated text to the WebSocket client
        async for new_text in streamer:
            await websocket.send_json({"token": new_text})

        # Notify the client that streaming is complete
        await websocket.send_json({"status": "Streaming completed."})
        await websocket.close()

    except Exception as e:
        await websocket.close(code=1011, reason=f"Error: {str(e)}")

# Expose the app using ngrok
if __name__ == "__main__":
    # Start ngrok with your static domain
    ngrok_tunnel = ngrok.connect(8000, bind_tls=True, hostname="ray-champion-crow.ngrok-free.app")
    print(f"Public URL: {ngrok_tunnel.public_url}")

    # Run the FastAPI application
    uvicorn.run(app, host="0.0.0.0", port=8000)
```
